<h1><u>3. Basic Insight of Dataset</u></h1><hr/>
      

### Dataset Information
This data set consists of three types of entities: (a) the specification of an auto in terms of various characteristics, (b) its assigned insurance risk rating, (c) its normalized losses in use as compared to other cars. The second rating corresponds to the degree to which the auto is more risky than its price indicates. Cars are initially assigned a risk factor symbol associated with its price. Then, if it is more risky (or less), this symbol is adjusted by moving it up (or down) the scale. Actuarians call this process "symboling". A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe.

The third factor is the relative average loss payment per insured vehicle year. This value is normalized for all autos within a particular size classification (two-door small, station wagons, sports/speciality, etc...), and represents the average loss per car per year.

Note: Several of the attributes in the database could be used as a "class" attribute.

<h2 id="identify_handle_missing_values"><u>4.1 Identify and handle missing values</u></h2>


<h3 id="identify_missing_values">Identify missing values</h3>
<h4>Convert "?" to NaN</h4>
In the car dataset, missing data comes with the question mark "?".
We <b>replace "?" with NaN </b>(Not a Number), which is Python's default missing value marker.

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

<h3>Evaluating for Missing Data</h3>

```python
missing_data = df.isnull()
missing_data.head()
```

The output is a boolean value and "True" stands for missing value, while "False" stands for not missing value..

<h3>Count missing values in each column</h3>

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

Based on the summary above, each column has 205 rows of data, <b><i>seven columns</b></i> containing missing data:
<ol>
    <li><b><i>"normalized-losses":</i></b> 41 missing data</li>
    <li><b><i>"num-of-doors":</i></b>  2 missing data</li>
    <li><b><i>"bore":</i></b>  4 missing data</li>
    <li><b><i>"stroke" :</i></b>  4 missing data</li>
    <li><b><i>"horsepower": </i></b> 2 missing data</li>
    <li><b><i>"peak-rpm":</i></b>  2 missing data</li>
    <li><b><i>"price":</i></b>  4 missing data</li>
</ol>

<p><b><font size = "3">Replace by mean:</font></b></p>
<ol>
    <li>normalized-losses</li>
    <li>bore</li>
    <li>stroke</li>
    <li>horsepower</li>
    <li>peak-rpm</li>
    
</ol>

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
avg_bore=df['bore'].astype('float').mean(axis=0)
avg_stroke=df['stroke'].astype('float').mean()
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)

df["normalized-losses"].replace(np.nan, int(avg_norm_loss), inplace=True)
df['bore'].replace(np.nan, round(avg_bore,2), inplace=True)
df['stroke'].replace(np.nan, round(avg_stroke,2), inplace=True)
df['horsepower'].replace(np.nan, int(avg_horsepower), inplace=True)
df['peak-rpm'].replace(np.nan, int(avg_peakrpm), inplace=True)
```

<p><b><font size = "3">Replace by frequency::</font></b></p>
<p><b>num-of-doors: </b>Because 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</p>
    

```python
df["num-of-doors"].replace(np.nan, df['num-of-doors'].value_counts().idxmax(), inplace=True)
```

<p><b><font size = "3">Drop:</font></b></p>
<p><font size = "2">Drop all rows that do not have price data</font></p>

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

<h2 id="identify_handle_missing_values"><u>4.2 Data Standardization</u></h2>
<p><b>Transform mpg to L/100km:</b></p>
<p>In our dataset, the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we are developing an application in a country that accept the fuel consumption with L/100km standard</p>

```python

df['city-mpg'] = round(df["city-mpg"]/2.35,2)
df['highway-mpg']=round(df['highway-mpg']/2.35,2)
df.rename(columns={'city-mpg':"city-km/L","highway-mpg":'highway-km/L'}, inplace=True)

df.head()
```


<h2 id="identify_handle_missing_values"><u>4.3 Data Normalization</u></h2>
<p> We are going to Normalize the columns "length", "width" and "height"  so their value ranges from 0 to 1.</p>


```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
df[['length', 'width', 'height']].head()
```
